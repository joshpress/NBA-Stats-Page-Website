```python
import json
import time
from nba_api.stats.endpoints import leaguegamelog
from nba_api.stats.library.parameters import SeasonType

def get_nba_games(start_year, end_year):
    all_games = []
    #filter for more than 10 games played
    
    for year in range(start_year, end_year):
        season_str = f"{year}-{str(year+1)[2:]}"
        print(f"Fetching {season_str}...")
        
        try:
            log = leaguegamelog.LeagueGameLog(season=season_str, season_type_all_star=SeasonType.regular)
            

            data_dict = log.get_dict()['resultSets'][0]
            headers = data_dict['headers']
            rows = data_dict['rowSet']
            

            for row in rows:
                game_entry = dict(zip(headers, row))
                all_games.append(game_entry)
                
            time.sleep(1)
        except Exception as e:
            print(f"Error on {season_str}: {e}")


    with open('nba_games_2004_2026.json', 'w') as f:
        json.dump(all_games, f, indent=4)

get_nba_games(2004, 2026)
```

```python
import json
import time
from nba_api.stats.endpoints import playergamelogs
from nba_api.stats.library.parameters import SeasonType

def get_player_stats(start_year, end_year):
    all_player_stats = []
    
    for year in range(start_year, end_year):
        season_str = f"{year}-{str(year+1)[2:]}"
        print(f"Fetching player stats for {season_str}...")
        
        try:
    
            log = playergamelogs.PlayerGameLogs(
                season_nullable=season_str,
                season_type_nullable=SeasonType.regular
            )
            
            data_dict = log.get_dict()['resultSets'][0]
            headers = data_dict['headers']
            rows = data_dict['rowSet']
            

            for row in rows:
                stat_entry = dict(zip(headers, row))
                #get rid of rank and fantasy
                clean_entry={k:v for k,v in stat_entry.items() if"_RANK" not in k and "FANTASY" not in k}
                clean_entry["GAME_DATE"] = clean_entry["GAME_DATE"].split("T")[0]
                all_player_stats.append(clean_entry)
                

            time.sleep(1.5)
            
        except Exception as e:
            print(f"Error fetching {season_str}: {e}")

 
    with open('nba_player_stats_2004_2026.json', 'w') as f:
        json.dump(all_player_stats, f, indent=4)
    
    print(f"Saved {len(all_player_stats)} player stat lines to JSON.")


get_player_stats(2004, 2026)
```
